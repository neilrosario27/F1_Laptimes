```python
response = urlopen('https://api.openf1.org/v1/laps?session_key=latest&driver_number=1&lap_number>=0')
data_max = json.loads(response.read().decode('utf-8'))
print(data_max)
```

```python
df = pd.DataFrame(data_max)
df_cleaned_max = df.drop(columns=['meeting_key', 'session_key', 'i1_speed', 'i2_speed', 'st_speed', 'duration_sector_1', 'duration_sector_2', 'duration_sector_3', 'segments_sector_1', 'segments_sector_2', 'segments_sector_3'])
```

```python
print(df_cleaned_max)
```

```python
import plotly.graph_objects as go


trace1 = go.Scatter(
    y=df_cleaned_osc['lap_duration'],
    mode='lines',
    name='OSC Lap Duration',
    line=dict(color='orange', dash='dot')
)

trace2 = go.Scatter(
    y=df_cleaned_lan['lap_duration'],
    mode='lines',
    name='LAN Lap Duration',
    line=dict(color='orange', dash='solid')
)

trace3 = go.Scatter(
    y=df_cleaned_max['lap_duration'],
    mode='lines',
    name='MAX Lap Duration',
    line=dict(color='blue', dash='solid')
)

fig = go.Figure()
fig.add_trace(trace1)
fig.add_trace(trace2)
fig.add_trace(trace3)

fig.update_layout(
    title='Lap Duration Comparison',
    xaxis_title='Index',
    yaxis_title='Lap Duration',
    legend_title='Legend'
)

fig.show()
```
